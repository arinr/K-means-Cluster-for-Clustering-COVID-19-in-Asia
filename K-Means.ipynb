```python
#Mengambil data hanya negara yang termasuk di benua Asia
df_asia=df[df["Negara Asia"]==1]
df_asia=df_asia[["Country","Totdeath1M"]]
df_asia
```

```python
#Membuang negara yang tidak memiliki data kematian
df_asia=df_asia.dropna()
df_asia
```

```python
#Mengambil kolom total death per 1 milion population
df_death=df_asia[["Totdeath1M"]]
df_death.set_index(df_asia["Country"],inplace=True)
df_death
```

```python
#Mengecek type data
df_death.dtypes
```

```python
#Menghapus koma pada data agar data bisa dirubah ke type float dari type object
df_death['Totdeath1M']=df_death['Totdeath1M'].str.replace(',','')
df_death
```

```python
#Mengubah type data objek menjadi float
df_death['Totdeath1M']=df_death['Totdeath1M'].astype(float)
df_death.dtypes
```

```python
#Menstandarisasi data
from sklearn import preprocessing
df_standar=pd.DataFrame(preprocessing.scale(df_death["Totdeath1M"]),columns=['Totdeath1M'])
df_standar.set_index(df_asia["Country"],inplace=True)
df_standar.head()
```

```python
#MembuatScatter plot dari Country dan Totdeathm1M
import matplotlib.pyplot as plt
plt.figure(figsize=(16,8))
plt.scatter(df_asia["Country"],df_death["Totdeath1M"])
plt.xticks(rotation=90)
plt.xlabel('Negara')
plt.ylabel('Standardize Totdeath1M')
plt.title('Scatter Plot Totdeath1M')
plt.savefig("Scatter Plot.png")
plt.show()
```

```python
#Memilih k-cluster yang optimal dengan metode Elbow
from sklearn.cluster import KMeans
distortions = []
K = range(1,10)
for k in K:
    kmeanModel = KMeans(init='k-means++',n_clusters=k)
    kmeanModel.fit(df_standar)
    distortions.append(kmeanModel.inertia_)
```

```python
#Memplotkan metode Elbow
plt.figure(figsize=(16,8))
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.savefig("Elbow Method.png")
plt.show()
```

```python
#Mengestimasi model K-means dengan 2 cluster
kmeanModel = KMeans(n_clusters=2,init='k-means++')
kmeanModel.fit(df_standar)
```

```python
#Memplotkan hasil clustering dengan 2 cluster
plt.figure(figsize=(16,8))
plt.scatter(df_asia["Country"],df_death["Totdeath1M"],c=kmeanModel.predict(df_standar))
plt.xticks(rotation=90)
plt.xlabel('Negara')
plt.ylabel('Standardize Totdeath1M')
plt.title('Scatter Plot Totdeath1M')
plt.savefig("Hasil Cluster.png")
plt.show()
```

```python
#Menambahkan kolom k-means yang isinya hasil prediksi cluster tiap negara di benua Asia
df_death['k-means']=kmeanModel.predict(df_standar)
df_death.head()
```

```python
#Membua file excel dari data frame
df_death.to_excel('Hasil k-means.xlsx')
```
